# 1) Department wise attendance of students

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('/home/amh/Program/Python/dataset/student_data_clean.csv')

dep_att = df.pivot_table(index="Department", values="Attendance_Rate", columns='Gender', aggfunc="mean")
dep_att.plot(kind='bar', figsize=(5, 5))
```

# 2) Gender wise number of students

```python
df["Gender"].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(5, 5), ylabel="")
plt.title("Gender Distribution")
plt.show()
```

# 3) CGPA distribution by gender

```python
df.groupby("Gender")["CGPA"].mean()
```
