# 1) Calculate total sales by region and find which region has the highest sales

```python
import pandas as pd

data = pd.read_csv("/home/amh/Program/Python/dataset/sales.csv")
df = pd.DataFrame(data)
df.groupby("region")["sales"].sum()
```

# 2) Find the average discount given for each product category

```python
df.groupby("product")["discount"].mean()
```

# 3) Analyze customer ratings - calculate average rating by customer type (New vs Returning)

```python
data2 = pd.read_csv("/home/amh/Program/Python/dataset/customers.csv")
df2 = pd.DataFrame(data2)
df_merged = pd.merge(df, df2, on="order_id")
df_merged.head(8)
df_merged.groupby("customer_type")["customer_rating"].mean()
```

# 4) Find top 5 products by quantity sold

```python
df_merged.groupby("product")["quantity"].sum().nlargest(5)
```

# 5) Calculate monthly sales trend by extracting month from date column (check pandas date time modules)

```python
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.month

df.groupby("month")["sales"].sum().reset_index()
```
