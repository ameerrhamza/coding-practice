# 1) Check for missing values in each column and handle them appropriately

```python
import pandas as pd

data = pd.read_csv("/home/amh/Program/Python/dataset/sales.csv")
df = pd.DataFrame(data)
df.isnull().sum()
```

# 2) Check for duplicate rows and remove any duplicates if found

```python
df.duplicated().sum()
```

# 3) Convert 'date' column to datetime format

```python
df["date"] = pd.to_datetime(df["date"])
df.info()
```

# 4) Check data types of all columns and convert if necessary (e.g., ensure 'quantity' is int, 'price' is float)

```python
df["price"] = df["price"].astype(float)
df.info()
```

# 5) Standardize 'region' values (ensure consistent capitalization and remove extra spaces)

```python
df["region"] = df["region"].str.capitalize().str.strip()
df.head(8)
```
